```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

Answer to the **Q1**

```python
import numpy as np
import pandas as pd

df = pd.read_csv("employees.csv")

ages = df["age"].to_numpy()
salaries = df["salary"].to_numpy()

print(ages)
print(salaries)

print(ages.dtype)
print(ages.ndim)
print(ages.shape)
print(ages.size)

print(salaries.dtype)
print(salaries.ndim)
print(salaries.shape)
print(salaries.size)
```

**Answer to the Q2**

```python
highest_sal = np.max(salaries)
lowest_sal = np.min(salaries)
avg_sal = np.mean(salaries)
avg_age = np.mean(ages)

print(highest_sal)
print(lowest_sal)
print(avg_sal)
print(avg_age)
```

**Answer to the Q3**

```python
age_over_30 = ages[ages>30]
count_over_30 = age_over_30.size

print(age_over_30)
print(count_over_30)
```
